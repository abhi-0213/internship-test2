```python
import pandas as pd
data = pd.read_csv("main.csv")
```

```python
data["sales_amount"]=data["sales_quantity"].where(data["unit"]!="pcs",
data["product_description"].str.split("-",expand=True)[1].astype("float")*\
data["sales_quantity"])
data["price_new"] = data['price'].fillna(
   data.groupby('product_description')['price'].transform("mean")
)
data.tail()
```

```python
data.to_csv("main.csv")
```
